```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime

```

```python

ds = pd.read_csv("udp_data_timestamp.txt", header=None, delimiter=',')
tp= ds.loc[:,0]

```

```python
port1= ds.loc[:,:14]
port1 = port1.rename(columns={0: 'timestamp',1: 'p1', 2: 'Port', 3:'R1_1', 4: 'Phi1_1', 5: 'R1_2', 6: 'Phi1_2', 7: 'R1_3', 8: 'Phi1_3', 9: 'R1_4', 10: 'Phi1_4', 11: 'R1_5', 12: 'Phi1_5', 13: 'R1_6', 14: 'Phi1_6'})
port1= port1.drop('p1', axis=1)
port1 = port1.drop('timestamp', axis=1)
port1
```

```python
port2 = ds.loc[:, 15:28]
port2 = port2.rename(columns={15:'p2' ,16: 'Port', 17:'R2_1', 18: 'Phi2_1', 19: 'R2_2', 20: 'Phi2_2', 21: 'R2_3', 22: 'Phi2_3', 23: 'R2_4', 24: 'Phi2_4',25:'R2_5', 26:'Phi2_5', 27:'R2_6', 28:'Phi2_6'})
port2= port2.drop('p2', axis=1)
port2
```

```python
port3= ds.loc[:,29:38]
port3 = port3.rename(columns={29:'p3' ,30: 'Port', 31:'R3_1', 32: 'Phi3_1', 33: 'R3_2', 34: 'Phi3_2', 35: 'R3_3', 36: 'Phi3_3', 37: 'R3_4', 38: 'Phi3_4'})
port3
```

```python


def reciprocal(df):
    # Identify numeric columns
    numeric_columns = df.select_dtypes(include=[np.number]).columns

    # Exclude 'Port' column
    numeric_columns = numeric_columns.drop('Port')
    #numeric_columns = numeric_columns.drop('timestamp')
    
    # Calculate the reciprocal for numeric columns
    df[numeric_columns] = df[numeric_columns].apply(lambda x: 1 / x)
    
    return df
```

```python
ds = reciprocal(ds)
ds
```

```python
ds.to_csv('data.csv', index=False)
```
